```python
#讀取csv檔成為2維的list
import csv
with open(f'a2.csv',encoding='utf-8', newline='') as file:
    csv_reader = csv.reader(file)   #左邊的名稱csv_reader是自己定的
    next(csv_reader)     #next是新概念!!
    #print(type(csv_reader))
    csv_list = list(csv_reader)
    display(csv_list)
```

```python
for students in csv_list:
    if students[0] == '陳信念':
        print(students)
        break
```
