```python
#使用codesapce 要安裝中文字型
#install 要加-y
#命令列:sudo apt-get update -y                         ←複製貼上 左邊那串命令列到終端機做下載
#命令列:sudo apt-get install -y fonts-wqy-zenhei       ←複製貼上 左邊那串命令列到終端機做下載
#說明網址:https://askubuntu.com/questions/490829/how-can-i-install-chinese-fonts-on-kubuntu-14-04
#字型位置:/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc       ←左邊fname~.ttf是字型的路徑

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

#mac font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)     ←左邊fname~.ttf是字型的路徑
#mac font1=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=12)   ←左邊fname~.ttf是字型的路徑
#windows font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc,size=8)
font=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)    
font1=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=12)

student1 = {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2 = {'國文':90, '英文':75, '數學':98, '歷史':65, '探索':75}
student1.keys()
student1.values()

figure = plt.figure(figsize=[8,5],facecolor='#cccccc')   #建立畫布
axes = figure.add_subplot(1,1,1)  #加入圖表
axes.plot(student1.keys(), student1.values(),'r--+',label='學生1')  #設定XY軸,標記的種類樣式,圖例標籤
axes.plot(student2.keys(), student2.values(),'b--.',label='學生2')  #設定XY軸 ,標記的種類樣式,圖例標籤
axes.set_xticklabels(student1.keys(),fontproperties=font)  # 本來中文跑不出來，用fontproperties=font來解決我選用上方size=8的變數
axes.set_title('學生1的分數',fontproperties=font)   #設定圖表標題
axes.set_xlabel('科目',fontproperties=font)    #設定X軸的標籤
axes.set_ylabel('分數',fontproperties=font)    #設定Y軸的標籤
axes.grid(axis='y')     #顯示y軸背景線
axes.legend(prop=font)  #設置圖例
plt.show()     #顯示圖表
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

#mac font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)     ←左邊fname~.ttf是字型的路徑
#mac font1=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=12)   ←左邊fname~.ttf是字型的路徑
#windows font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc,size=8)
font=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)    
font1=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=12)

student1 = {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2 = {'國文':90, '英文':75, '數學':98, '歷史':65, '探索':75}
student1.keys()
student1.values()

figure = plt.figure(figsize=[8,5],facecolor='#cccccc')   #建立畫布
axes = figure.add_subplot(1,1,1)  #加入圖表
student1_values_list  = list(student1.values())   #把value(dict)轉成 list
student2_values_list  = list(student2.values())   #把value(dict)轉成 list 
axes.plot(student1.keys(),student1_values_list,'r--+',label='學生1')  #設定XY軸,標記的種類樣式,圖例標籤
axes.plot(student2.keys(),student2_values_list,'b--.',label='學生2')  #設定XY軸 ,標記的種類樣式,圖例標籤
axes.set_xticklabels(student1.keys(),fontproperties=font)  # 本來中文跑不出來，用fontproperties=font來解決我選用上方size=8的變數
axes.set_title('學生1的分數',fontproperties=font)   #設定圖表標題
axes.set_xlabel('科目',fontproperties=font)    #設定X軸的標籤
axes.set_ylabel('分數',fontproperties=font)    #設定Y軸的標籤
axes.grid(axis='y')     #顯示y軸背景線
axes.legend(prop=font)  #設置圖例
#axes.annotate('75',[1,90])    #annotate=註解
xlocation =axes.get_xticks()    #用get_xticks()得到X軸的[0,1,2,3,4]座標(位置)，類似list
#ylocation =axes.get_yticks()
xtick_count = len(xlocation)    # xtick_count的長度為5
for i in range(xtick_count):         #用迴圈為0,1,2,3,4
    x = xlocation[i]
    y1 = student1_values_list[i]
    y2 = student2_values_list[i]
    
    axes.annotate(str(y1),[x,y1])   #設定註解座標((註解內容),[X,Y座標])
    axes.annotate(str(y2),[x,y2])   #設定註解座標((註解內容),[X,Y座標])



plt.show()     #顯示圖表
```
