```python
#49:09
#複製17_6
import pandas as pd
df = pd.read_csv('學生成績.csv')  #讀進csv
df.index = df['學號']     #取學號欄的內容,覆蓋原本的索引
df1 = df.drop(columns='學號')  #移除學號欄
df1
```

```python
#53:39
df1['排名'] = df1['平均'].rank(method='min',ascending=False) #ascending=True表示升階,=False表示降階
df2 = df1.sort_values(by='平均',ascending=False)
```
