```python
#1:36:02~1:47:55
import yfinance as yf
data =yf.download("2330.TW", start="2022-01-01",end="2022-12-31")
tw2330 = data = data[["Adj Close"]]  #2:03:16   #Adj Close收盤價
tw2330.columns = ["台積電"]
tw2330
```

```python
#複製上面的 2:02:20~2:04:41
data =yf.download("2303.TW", start="2022-01-01",end="2022-12-31")
tw2303 = data = data[["Adj Close"]]  #2:03:16   #Adj Close收盤價
tw2303.columns = ["聯電"]
tw2303
```

```python
#複製上面的 2:04:50~2:07:18
data =yf.download("2454.TW", start="2022-01-01",end="2022-12-31")
tw2454 = data = data[["Adj Close"]]  #2:03:16   #Adj Close收盤價
tw2454.columns = ["聯發科"]
tw2454
```

```python
#複製上面的 2:07:19~2:08:24
data =yf.download("2317.TW", start="2022-01-01",end="2022-12-31")
tw2317 = data = data[["Adj Close"]]  #2:03:16   #Adj Close收盤價
tw2317.columns = ["鴻海"]
tw2317
```

```python
#2:08:26~2:11:00
import pandas as pd
frame=pd.concat([tw2330,tw2303,tw2454,tw2317],axis=1) #把上面四個資料合併起來 #2:15:36
frame
```

```python
#2:16:30~2:19:05
frame1 = frame.pct_change()    # pct_chang()顯示每筆資料間漲跌的比例
frame2 = frame1.fillna(0)    #用fillna(0)把Nan變成0
frame2
```

```python
#2:22:48
frame2.corr() #各股票的線性規劃
```
