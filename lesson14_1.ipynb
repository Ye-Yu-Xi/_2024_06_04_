```python
#寫程式下載檔案

import requests

url ="https://opendata.cwa.gov.tw/fileapi/v1/opendataapi/F-C0032-001?Authorization=rdec-key-123-45678-011121314&format=JSON"
r = requests.get(url)
if r.status_code == 200:                 
    print("下載成功")
#上面結果顯示200 代表連線成功、下載成功   # https://umbraco.com/knowledge-base/http-status-codes/
```

```python
print(r.text)
```

```python
#json特性: 1.純文字 2.複雜的結構
# https://jsonviewer.stack.hu/
all_json = r.json()
locations = all_json['cwaopendata']['dataset']['location']
for item in locations:
    print(item['locationName'])
```
